```python
records = []

def calculate_grade(percentage):
    """Return grade based on percentage."""
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "Fail"

def add():
    try:
        name = input("Enter the Student name: ").title()
        if not name.replace(" ", "").isalpha():
            print("‚ùå Name should contain only letters.\n")
            return

        roll_input = input(f"Enter the roll number for {name}: ")
        if not roll_input.isdigit():
            print("‚ùå Roll number must be numeric.\n")
            return
        roll_no = int(roll_input)

        subjects = ["Hindi", "English", "Maths", "Science", "Social Studies"]
        marks = {}

        print("\nEnter marks for each subject:")
        for subject in subjects:
            while True:
                mark = input(f"{subject}: ")
                if mark.isdigit():
                    marks[subject] = int(mark)
                    break
                else:
                    print("‚ùå Please enter numeric marks.")

        # Calculate total and percentage
        total = sum(marks.values())
        percentage = total / len(subjects)
        grade = calculate_grade(percentage)

        record = {
            "name": name,
            "roll_no": roll_no,
            "marks": marks,
            "total": total,
            "percentage": round(percentage, 2),
            "grade": grade
        }

        records.append(record)

    except Exception as e:
        print(f"‚ö†Ô∏è An unexpected error occurred: {e}")
    else:
        print(f"\n‚úÖ Student {name}'s record added successfully!\n")

def view():
    if not records:
        print("üìÇ No student records available.\n")
    else:
        print("\n===== All Student Records =====")
        for r in records:
            print(f"\nName: {r['name']}, Roll No: {r['roll_no']}")
            print("Marks:")
            for subject, mark in r['marks'].items():
                print(f"  {subject}: {mark}")
            print(f"Total Marks: {r['total']}")
            print(f"Percentage: {r['percentage']}%")
            print(f"Grade: {r['grade']}")
        print("\n===============================\n")

def delete():
    try:
        roll = input("Enter roll number to delete: ")

        if not roll.isdigit():
            print("‚ùå Roll number should be numeric.\n")
            return

        roll_no = int(roll)

        for r in records:
            if r['roll_no'] == roll_no:
                records.remove(r)
                print("‚úÖ Record deleted successfully!\n")
                break
        else:
            print("‚ùå No record found with that roll number.\n")

    except Exception as e:
        print(f"‚ö†Ô∏è An unexpected error occurred: {e}")

while True:
    print("""
    ===== Student Record System =====
    1. Add a student
    2. View all students
    3. Delete a student
    4. Exit
    =================================
    """)

    try:
        number = int(input("Enter your choice (1‚Äì4): "))
    except ValueError:
        print("‚ùå Please enter a valid number between 1 and 4.\n")
        continue

    if number == 1:
        add()
    elif number == 2:
        view()
    elif number == 3:
        delete()
    elif number == 4:
        print("üëã Exiting program. Goodbye!")
        break
    else:
        print("‚ùå Invalid choice! Please enter a valid number.\n")
```
